Now we shall define some functions to create Histograms and KDE functions for any particular column of the dataframe

```python
def plot_hist(field_name,bins=50):
    args={"legend":True,"figsize":(8,8),"kind":"hist","title":"Probability Distribition for "+field_name,"bins":bins}
    df_original[field_name][df_original["class"]==0].plot(label="Not Fraud",**args)
    df_original[field_name][df_original["class"]==1].plot(label="Fraud",**args)
    
def plot_kde(field_name):
    args={"legend":True,"figsize":(8,8),"kind":"kde","title":"Probability Distribition for "+field_name}
    df_original[field_name][df_original["class"]==0].plot(label="Not Fraud",**args)
    df_original[field_name][df_original["class"]==1].plot(label="Fraud",**args)    
```

Example output for Purchase value field. Now from this KDE function we can see that the Fraud and Non-Fraud classes have very similar distributions for Purchase value. Now we get on to the problem of mapping the IP address to a country. We first read the csv file and store it with the lower bound as the index (this will help us in mapping later)  

```python
df_ip_add=pd.read_csv("IpAddress_to_Country.csv",index_col=0)
df_ip_add.head()
```

Now we we use the ip adress in the original data frame to get the country. We do this by finding the index of the ip_address dataframe which is immediately before (by using the pad keyword) the ip address we pass to the function below. Since we used the lower bound as the index, we get back the row with the correct ip address. Then we just return the country for that particular row. Note that some of the entries in the dataframe were throwing errors, because the elements we not in the list below, so we had to use exception handling and return "No Value" for those values . 

```python
def get_country(x):
    try:
        return df_ip_add.iloc[df_ip_add.index.get_loc(np.round(x),method='pad')]["country"]
    except:
        
        return "No value"
```

Now we apply this function to the ip_address column and store it as a new column "Country"

```python
df_original["Country"]=df_original["ip_address"].apply(get_country)
```

"Country" will now be a categorical variable with 182 values. This kind of categorical variable is hard to handle. So I decided to bin the countries in terms of frequency of occurance. The most frequent to least frequent countries are put in bins 0 to 2 , with the "No value" entries put in bin -1. These could be imputed later. The threshholds for the bins are 10,000, 500 and 0. I tried other binnings as well and found that it does not really change things.  

```python
country_val_count=df_original[df_original["Country"].notnull()]["Country"].value_counts()
```

```python
def frequency_mask(x):
    if x=='No value':
        return -1
    if country_val_count[x]> 10000:
        return 0
    elif country_val_count[x] >500:
        return 1
    else:
        return 2
    
```

Using this binning function to create the binned Country colun called "Country_mask"

```python
df_original["Country_mask"]=df_original["Country"].apply(frequency_mask)
```

Checking bins formed and also the most common countries 

```python
print df_original["Country_mask"].value_counts()
df_original["Country"].value_counts().head()
```

Label Encoding the other categorical variables such as Source , browser and sex

Now we come to the problem of how to use the device_id field . It struck me that fraudsters would probably be reusing the same device and making a lot purchases. So we I decided to mark the use count for each individual device id. This is stroed in the column "device use frequency"

```python
freq_use_dict=df_original["device_id"].value_counts()
df_original["device use frequency"]=df_original["device_id"].apply(lambda x: freq_use_dict[x])
```

The only other columns which need parsing are the signup time and purchase time. Again it struck me that if I were a fraudster I would not really create an account a long time before I purchased things. Therefore I decided to create a new variable called "hours_to_purchase" which calculates the number of hours passed between signup and purchase.  

Note that we do not do anything to the user_id because it is unique for every entry and therefore not useful.In fact if you pass the user id column to a massive Decision tree it would probably create singletons by splitting on it, which is pointless because it will not be useful on test data. Ofcourse there are ways to get around this by using Information gain ratio or by pruning it intelligently, but that is besides the point.

```python
df_original["hours_to_purchase"]=\
((pd.to_datetime(df_original["purchase_time"])-pd.to_datetime(df_original["signup_time"]))).astype('timedelta64[s]')/3600
```

# Lets start with logistic regression

Let us start off with the old faithful, Logistic Regression. First off we load some packages from sklearn. We will be using GridsearchCV to to optimize out classifier over hyperparameters by using K-Fold cross-validation. We will also be using a pipeline because we want to see what effects Standard Scaling has on our classifer. Finally we will also be using the train_test module to create training and test sets

```python
from sklearn.grid_search import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import train_test_split
```

Lets start with the most promising fields-hours to purchase,country mask(maybe?) and device use frequency. We convert the values in the pandas dataframe to a numpy array because that is what sklearn likes. Note that there is a pandas+sklearn library as well which can do things directly on dataframes. However I do not have that installed so we will be converting to numpy arrays using the values member variable of a pandas series.

```python
to_keep_ml=["device use frequency","hours_to_purchase","Country_mask"]
x_array=df_ml[to_keep_ml].values
y_array=df_ml["class"]
```

Here we see that the X (feature set) and Y( target) have the correct shapes

```python
x_array.shape,y_array.shape
```

Now we use Train Test Split to split the data in training and test sets. Note that we use a test set that is a quarter of the actual data set. We use Stratified split, which maintains the relative proportions in the two sets. We also fix the random state as we want to get some reproducibilty when we redo this step in the future

```python
Train_X,Test_X,Train_Y,Test_y=train_test_split(x_array,y_array,test_size=0.25,stratify=y_array,random_state=42)
```

Checking the class proportions of the Test set. I did this to basically check that the stratified splitting  worked as expected 

```python
from collections import Counter
Counter(Train_Y)
```

Now we use Pipeline to set up a pipeline using standard scaler and Logistic regression. Note that we again set the random state and later we shall also test whether Standard scaler being on/off works better.

```python
Pipe_Lr=Pipeline([("Scaler",StandardScaler()),("LR",LogisticRegression(random_state=84))])
```

We choose a list to create the list of regularization parameters for the regularization parameter in log series. 

```python
List_C=np.logspace(-5,5,11)
print List_C
```

Now we get to the parameter grid. If you have used spark.ml library in python, this should be familiar. Basically we create a parameter grid for different hyperparameters. Note that when we define the PipeLine, the different steps are given names (LR for Logistic Regression) in the tuples we define. Then we can specify a hyperparameter by using the syntax name__hyperparameter={Dictionary of values}.

Note that on my macbook air due to limited resources, I cannot do a single gridsearch over all possible things I want to search over as the process takes too long and I have to wait a long time without having any idea what is going on. Therefore in the pipelines in this book you see in this book from here on, you will see small grids. I however have tried many many more combinations by refining the search many times over. What you see has been attained after many other things were tried (in most cases)

Now let us get to the things we see here. The "with_std" parameter for scaler decides if we want to normalize to an std of 1.  The C I have already spoken about (regularization parameter). The penalty parameter decides whether to use L2 or L1 regularization.The most important parameter is the class weight. Now we want to over-represent the fraud class because we want that to be predicted correctly. The higher the weight, the more the classifier prioritizes it. Note sklearn also has a "balanced" mode which sets class weights to inverse of the occurance frequencies. 



```python
Param_grid_Lr={"Scaler__with_std"
               :[True,False],"LR__C":List_C,"LR__penalty":['l1','l2'],\
               "LR__class_weight":["balanced",{0:1,1:1},{0:1,1:2},{0:1,1:5}]}
```

In this step we now use the grid search cross validation to optimize our classifier. We perform 4 fold stratified cross validation and use the F1 score of the fraud class as our optimization parameter. We choose this because right now this seems like a reasonable tradeoff between precision and recall (optimizing either one completely would lead to almost dummy classifiers in the extreme limit).

We then use the best classifer to fit to the training set and compute metrics on the test set.

```python
best_Lr=GridSearchCV(Pipe_Lr,Param_grid_Lr,cv=4,verbose=1,n_jobs=1,scoring='f1')
```

```python
best_Lr.fit(Train_X,Train_Y)
```

```python
best_Lr.best_params_,best_Lr.best_score_
```

```python
chosen_Lr=best_Lr.best_estimator_

chosen_Lr.fit(Train_X,Train_Y)
choser_Lr_train=chosen_Lr.predict(Train_X)
```

```python
from sklearn.metrics import classification_report
```

We now output the precision and recall and f1 scores for the training set to see how we did on the training set. As we see our recall is very low while our precision is high, which means we have too "Bold" a classifier. 

```python
print classification_report(Train_Y,choser_Lr_train)
```

We now output the coefficients of the logisitic regression classifier for the three fields. As we see , the hours to purchase (1st coefficient) and the use frequency (second coefficient) have significant values while country mask seems like a redundant parameter as expected.  The signs also make sense because of the depencies of frauds on these two variables (positive correlation and negative correlation respectively)

```python
chosen_Lr.named_steps["LR"].coef_
```

Checking metrics on the Test set. We see that we almost get the same metrics as the training set. Which means that we did not overfit.

```python

choser_Lr_test=chosen_Lr.predict(Test_X)
print classification_report(Test_y,choser_Lr_test)
```

Now we compute the ROC curve and the ROC AUC score for this classifier. We get an AUC score of around 0.83 which is rather reasonable since we have taken rather few features.

```python
from sklearn.metrics import roc_curve,auc
```

```python
y_dec_func=chosen_Lr.decision_function(Test_X)
fpr_lr,tpr_lr,thresholds_log=roc_curve(Test_y,y_dec_func)

print "AUC for Logistic Regression={}".format(auc(fpr_lr,tpr_lr))

plt.figure(figsize=(6,6))
plt.plot(fpr_lr, tpr_lr,'r-',label="Logistic Regression")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc=4)
```

# Lets redo things in a way that better simulates real-time data

Now at this point, which was late last night, I could have gone on and done more feature engineering to optimize this earlier scheme more. However I realized something important. What we are doing is not completely correct for the real world real time data this classifier will most probably be used for. For example in the device use frequency field, we find the number of uses for that particular device over all time and assign that value to all the entries with that device id. This is not something that we can do in real life because if we get this data in real time, we cannot possibly know how many times the person will use the device in the future. So we are picking up a lot of first time frauds which we should not be picking up. 

So the better model the real time case, we need to first sort the data frame in order of purchase time. Then we calculate for each entry, the number of times that device has been used in the past, the number of times that device has made a fraudulent transaction in the past and the number times a person with the same age has logged on from the same device.  Note that we could also do this for the other categorical variables but I only did this for age as of now because calculating these things takes a lot of time on my laptop (more on this later) and what i wanted to use the age as was really a "same user identifier" because for some reason fraudsters kept putting in the same age/sex/purchase value for the same device when committing fraud. In the actual polished version of the final model, we would treat a combination of age+purchase value+ sex+source etc as a user identifier and could use that to mark people.  



## Reforming data frame by sorting things in order of purchase time and calculating variables only for time in the past

```python
df2=df_original.copy()
df2["purchase_time"]=pd.to_datetime(df2["purchase_time"])
df2["signup_time"]=pd.to_datetime(df2["signup_time"])
df_sorted=df_original.sort_values("purchase_time")
```

Now we try to create these variables for the number of previous frauds, number of previous transactions with the same age and numbers of times the device has been previously used. Now this is a process which is rather slow because for each entry we have to scan all the entries before it to compute this. This is order N^2 and therefore it takes time. Now ofcourse you could improve this by created hashing functions (a list of dictionaries or maybe using groupby operations or something) which would exchange time for space complexity. However in the real world, we would only be doing a lookup for a new value as it came along, and that would be order N, which takes milliseconds and so is probably not worth optimizing over. 

So I ran the functions for calculating these variables on different parallel notebooks and obtained the results and put them into the dataframe. The codes used for number of previous frauds and previous device use number is given in the two cells below. I would not recommend running this as it too of the order of 10 mins to get these variables.

## Illustrative example showing how we could optimize for recall instead to build a more conservative classifier

In this case we would increase the weight of the fraud class by a lot. The more you try to increase the weight, the more the classifier tries to avoid false negatives for the fraud class. Now this ofcourse leads to a lot of false positives as well, which lowers the precision. In the example below we have made the class weights 100:1 (this is ofcourse far too high but we are just trying to make a point) to illustrate this effect. As you can see, the recall is almost 1 now, but precision has majorly suffered.

```python
NPipe_Rf_recall=Pipeline([("Scaler",StandardScaler()),("Rf",RandomForestClassifier())])
NParam_grid_Rf_recall={"Scaler__with_std":[False],'Rf__n_estimators':[25],'Rf__min_samples_split':[500],\
                  'Rf__max_features':['auto'],'Rf__class_weight':[{0:1,1:100}]}
Nbest_Rf_recall=GridSearchCV(NPipe_Rf_recall,NParam_grid_Rf_recall,cv=4,verbose=1,n_jobs=1,scoring='recall')
Nbest_Rf_recall.fit(NTrain_X,NTrain_Y)
print Nbest_Rf_recall.best_params_,Nbest_Rf_recall.best_score_
```

```python
Nchosen_Rf_recall=Nbest_Rf_recall.best_estimator_
Nchosen_Rf_recall.fit(NTrain_X,NTrain_Y)
Nchosen_Rf_recall_train=Nchosen_Rf_recall.predict(NTrain_X)
Nchosen_Rf_recall_test=Nchosen_Rf_recall.predict(NTest_X)
print classification_report(NTest_y,Nchosen_Rf_recall_test)
```

# Ideas for the future

At this point I believe that I can list future ideas to make the classifier better. All of these are implementable within a couple of hours each (coding+testing) and in principle might lead to better classification. Some of these points are

1) One of the clear trends that was spotted was that people who committed frauds make a lot of purchases with the exact same purchase value/sex/age/source/browser etc. So what we should really do is create some sort of a variable which tells us whether a user with the exact same values of these variables has logged in before and how many times. The variable would be similarly constructed to the age_before field. In a real world app, this field would be a preliminary red flag as multiple users with the exact same details are hard to imagine from the same device id.

2) The time to purchase field could also be reduced in range by binning all the high value entires in one large rather than keeping the whole range. This would reduce the dynamic range and might help things be more granular when standard scaler is used.

3) The purchase time could be used to calculate the current time of day in that country. If we get a very high number of logins from the same country with very few logins in the past with the same device id at an absurd time of day then we could probably mark them as fraud.

4) We could also do a simulation of a different train test split with the test set being the later time data instead of random samples taken all over. This might be useful for getting more true values of real world performance.

5) We also could do the binning of the countries differently. We could order them on fraud probability from that particular country. However that might run into issues of countries with low usage having skewed statistics.

6) We can ofcourse do other classifiers such as SVC. I actually tried doing SVC but for some reason it was incredibly slow on my computer. This might allow us to pick out non-linear decision boundried with higher order kernels which might be very accurate. 

7) We could at the end of the day train multiple classifiers and then develop a voting classifier which would pick by majority rule. This might be useful as in general different classifiers are good at different kinds of decision boundaries.

8) In some cases, the lack of computational resources hampered me. Cross validation and feature extraction are much easier if I would have something more powerful than my computer. In the real world I assume we would have access to more powerful machines where we could try out many more non-linear combinations of the features considered.

### Explain how different assumptions about the cost of false positives vs false negatives would impact the model. 

Ans- I think I have touched upon this earlier but I shall try to sum things up below. The main idea is that you can pick which side of the spectrum to err on. If you were to optimize for false positives then your false negatives would go up and vice versa. This is also known as the precision/recall trade off, because the two are usually negatively correlated . Now I have used the f1 score in most cases as my optimization objective, which is the harmonic mean of precision and recall. This aims to achieve a balance between the two. However in the case that you want to err on the side of caution, you would want something with high recall. This would lead to a lot of users who have genuine purchases being marked as fraud. Now this saves you from losing money to fraud but a website which harasses users needlessly by marking them as fraud would get very bad word of mouth which might lead to a dip in popularity. Therefore one must not go overboard in these things and what the particular sweet spot is for you depends on what your priorities are.


### How would you explain her how the model is making the predictions? Not from a mathematical perspective (she couldn't care less about that), but from a user perspective. What kinds of users are more likely to be classified as at risk? What are their characteristics? 

Ans- As of now the model is trying to make predictions based on a a few markers that are seen to be good indicators of fraud. 

1)The first is if the person makes a purchase very soon after after he signs up. Fraudulent robot programs tend to do this and it would be hard in principle for a human to make a purchase one second after he first signed up.

2) We also check the number of times the same device has been used beofre. Fraud robots just make similar accounts from the same device so devices which have been used before more are seen to have high probability of fraud.

3) We also check some other features such as the number of times somebody with the same age has logged in before and the number of times the number of frauds that have been committed from the same device. This all exploits the fact that fruad robots seem to create the same account type multiple times to conduct fraud.

With all the these features, we try to come up with a method that tries to catch fraud in a way that does not err too much on the side of caution nor is too bold. we attain an optimal state which balances these two factors (this is my layman's explanation for the f1 score) and checks for fraud. However by slight tuning of the parameters the method might be tuned so that we err more on either side.

People who fulfill the criteria above would be more at risk of being flagged. However I believe that the model is fairly good at picking out fraud robots. But if a smart human being were to commit fraud by changing his signup details every time he makes a new account and waiting a while before making the purchase then we might have to do some further feature engineering. 

### From a product perspective, how would you use it? That is, what kind of different user experiences would you build based on the model output?

As mentioned earlier , I have tried to tweak the model so that it is more conducive to real time processing. in this case, we would take purchse data as it comes in, calculate the variables mentioned above and then run the classifier. If the classifier outputs 1 (or equivalently we could set thresholds for the probabilities of fraud as predicted by the predict_proba() function of scikit learn), we would stop that purchase from going through. Note that we might just block the creation of accounts from a certain device id if that has a lot of fraudulent transactions in the past and we could store the list of these devices somewhere. If new accounts were to be created from these devices we could ask for higher levels of authentification such as cvv codes for credit cards, use captchas to confirm human purchases etc. In principle if we do detect high probability of fraud, we could flag the person at signup time itself if people with similar login details have committed frauds. If one wanted to still wanted to use such a flagged account we could again impose additional requirements on them like cvv/captcha etc. However it might be useful to not tell them that these steps are being undertaken because they have been flagged as probable fraudsters, because that might be seriously upsetting for genuine users (especially in the high recall case). We would just have an additional page show up with these requirements after they try to signup/purchase which says "For your protection, please complete the fields below". Note that we also require a little time for the model to run and predict things. We should not allow purchases to go through before the model has done its job.
