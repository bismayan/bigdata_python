```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("Saved_df.df",header=0,index_col=0)
df["signup_time"]=pd.to_datetime(df["signup_time"])
df["purchase_time"]=pd.to_datetime(df["purchase_time"])
```

```python
df_sorted=df.sort_values("purchase_time")
```

```python
from tqdm import tqdm_notebook,tnrange
fraud_series=[]
#for i in np.arange(np.arange(len(df_sorted))):
for i in tnrange((len(df_sorted))):
    temp=df_sorted.iloc[0:i+1]
    try:    
        fraud_series.append(temp[temp["device_id"]==df_sorted.iloc[i]["device_id"]]["class"].value_counts()[1])
    except KeyError:
        fraud_series.append(0)
```

```python
age_size=df_sorted["age"].max()-df_sorted["age"].min()
print age_size
print len(df["device_id"].value_counts().keys())
```

```python
from tqdm import tqdm_notebook,tnrange
age_series=[]
#for i in np.arange(np.arange(len(df_sorted))):
for i in tnrange((len(df_sorted))):
    temp=df_sorted.iloc[0:i+1]
    try:    
        age_series.append(temp[temp["device_id"]==df_sorted.iloc[i]["device_id"]]["age"].value_counts()[df_sorted.iloc[i]["age"]])
    except KeyError:
        age_series.append(0)
```

```python
fraud_before=[i-1 for i in fraud_series]
```
